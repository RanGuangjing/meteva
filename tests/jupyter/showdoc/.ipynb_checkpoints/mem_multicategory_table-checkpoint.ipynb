**<font face="黑体" color=blue size = 5>列联表</font>**  
**contingency_table_multicategory(ob, fo,grade_list=None,  member_list=None, save_path=None)**  
统计观测和预报分别为各种类别时对应的样本数  
  
|参数|说明|    
|:--|:--|  
|**<font face="黑体" color=blue size = 5>ob</font>**|实况数据, 任意维numpy数组|   
|**<font face="黑体" color=blue size = 5>fo</font>**| fo比Ob.shape多一维或者保持一致，fo.shape低维与ob.shape保持一致|   
|**grade_list**| 如果该参数为None，观测或预报值出现过的值都作为分类标记，如果该参数不为None，它必须是一个从小到大排列的实数，以其中列出的数值划分出的多个区间作为分类标签。对于预报和观测值不为整数的情况，grade_list 不能设置为None|  
|**member_list**|个预报成员的名称列表|     
|**save_path**| 如果该参数不为None,则混淆矩阵将会以excel文件形式保存到指定文件 |       
|<font face="黑体" color=blue size=5>return</font>|如果Fo和Ob的shape一致（即只有一种预报），返回内容为混淆矩阵的二维数组，shape为（类别数 + 1）×（类别数 + 1），如果Fo比Ob高出一维，则返回3维数组，shape为预报种类数据 ×（类别数 + 1）×（类别数 + 1）|  
 
 
**调用示例：**  

```python
#单个预报成员的检验，不设置等级参数的情况下，默认每一种数值都代表一种类别
result= mem.contingency_table_multicategory(ob,fo[0,:],
                save_path = r"H:\test_data\output\method\multi_catagory\ct1.xlsx")
```

```python
result.shape  #当只有一个预报成员时返回结果的shape
```

运行结果excel表格的截图将是如下样式

```python
#预报比观测高出一维时，返回3维数组
result= mem.contingency_table_multicategory(ob,fo,member_list = ["GRAPES","ECMWF"],
                    save_path = r"H:\test_data\output\method\multi_catagory\ct2.xlsx")
```

```python
result.shape
```

运行结果excel表格的截图将是如下样式,其中包含两个sheet部分GRAPES和ECMWF：

```python
result = mem.contingency_table_multicategory(ob,fo,grade_list=grade_list,member_list = ["GRAPES","ECMWF"],
                            save_path = r"H:\test_data\output\method\multi_catagory\ct3.xlsx")
```

```python
result.shape
```

**<font face="黑体" color=blue size = 5>频率表</font>**  
**frequency_table(ob,fo,grade_list=None,member_list = None, save_path=None)**  
统计观测和预报分别为各种类别的样本数  

  
|参数|说明|    
|:--|:--|  
|**<font face="黑体" color=blue size = 5>ob</font>**|实况数据, 任意维numpy数组|   
|**<font face="黑体" color=blue size = 5>fo</font>**| 比Ob.shape多一维或者保持一致，Fo.shape低维与ob.shape保持一致| 
|**grade_list**|阈值列表|  
|**member_list**|个预报成员的名称列表|     
|**save_path**| 如果该参数不为None,则混淆矩阵将会以excel文件形式保存到指定文件 |       
|<font face="黑体" color=blue size=5>return</font>|返回numpy数组，shape为（预报种类 + 1）×（类别数 + 1）|   
**调用示例：**  

```python
result = mem.frequency_table(ob,fo,grade_list=grade_list,member_list = ["GRAPES","ECMWF"],
                    save_path = r"H:\test_data\output\method\multi_catagory\ct4.xlsx")
```
